```python
# Importing libraries
import pandas as pd
import numpy as np
```

```python
# Reading olist_order_items_dataset.csv file
order_items = pd.read_csv('olist_order_items_dataset.csv')
order_items.head()
```

```python
# To get the column names
order_items.columns
```

```python
# Reading olist_orders_dataset.csv file
orders = pd.read_csv('olist_orders_dataset.csv')
orders.head()
```

```python
# To get the column names
orders.columns
```

```python
# Reading olist_order_reviews_dataset.csv file
review = pd.read_csv('olist_order_reviews_dataset.csv')
review.head()
```

```python
# To get the column names
review.columns
```

```python
# Merge order_items and orders dataset based on order_id and merged dataset to review dataset on order_id
df = pd.merge(order_items, orders, on='order_id',how = 'inner')
final_df = pd.merge(df, review[['order_id','review_score']],on = 'order_id',how = 'left')
final_df.head()
```

```python
# final dataFrame columns
final_df.columns
```

```python
# Dropping useless columns and Renaming columns
final_df = final_df[['product_id', 'customer_id', 'order_item_id', 'review_score']]
final_df = final_df.rename(columns={
    'product_id': 'Product',
    'customer_id': 'User',
    'order_item_id': 'Order_Quantity',
    'review_score': 'Review'
})
final_df.head()
```

```python
# To check null values
final_df.isnull().sum()
```

```python
# To handle null values
import warnings
warnings.filterwarnings('ignore')
final_df['Review'].fillna(final_df['Review'].median(),inplace=True)
final_df.isnull().sum()
```

```python
final_df.head()
```

```python
# Visualizing Top selling Products
top_products = final_df.groupby('Product')['Order_Quantity'].sum().sort_values(ascending=False).head(10)

# Visualizing using bar Chart
import matplotlib.pyplot as plt
top_products.plot(kind='bar', figsize=(10, 6), title='Top 10 Selling Products',color = 'g')
plt.xlabel("Product ID")
plt.ylabel("Total Quantity Sold")
plt.tight_layout()
plt.savefig('top_products.png')
plt.show()
```

This bar chart displays the top 10 best-selling products based on total quantity sold. Here’s what we can conclude:

High Performing Product:

The product with ID 422879e104f6682990de24d770e7f83d is the top-selling item, significantly outperforming all others with close to 800 units sold.

This product stands out with a clear lead over the rest, indicating it is likely a customer favorite or a high-demand item.

Moderate Range:

Products ranked 2 to 6 (e.g., IDs starting with aca2eb7, 3686c7, 53795a, etc.) show a moderate but consistent sales performance, ranging between 500–650 units.

These could represent steady-performing items, possibly essentials or well-marketed products.

Lower End of Top 10:

Products in the bottom 4 of the top 10 (starting with 706da07, a6225e, etc.) show comparatively lower sales, in the range of 300–380 units.

While still in the top 10, these may represent niche items or newer products gaining traction.

Skewed Distribution:

The distribution is right-skewed, with one product clearly dominating and others tapering down. This could suggest a bestseller effect or possibly limited competition in a certain category.

```python
# Visualizing Order Quantity Trends over time using line chart
final_df.groupby(final_df.index)['Order_Quantity'].sum().plot(kind='line', figsize=(10, 6), title='Order Quantity Trend Over Time')
plt.xlabel("Order Index")
plt.ylabel("Order Quantity")
plt.tight_layout()
plt.savefig('order_quantity_trend.png')
plt.show()
```

📊 Conclusion: Order Quantity Trend Over Time
This line plot visualizes the order quantity trend over time, where the x-axis represents the order index (chronological order of transactions), and the y-axis shows the quantity ordered in each transaction.

🧠 Key Observations:
Consistent Base Demand:

The majority of orders hover around quantities of 2 to 6 units, indicating that this is the typical purchase size for most customers.

This suggests steady consumer buying behavior over time.

High-Quantity Spikes:

Several distinct spikes above quantity 10 and even reaching above 20 are visible throughout the timeline.

These could be:

Bulk purchases by B2B customers

Promotional or discount-driven orders

Stocking or restocking behaviors

No Strong Seasonal Pattern:

There’s no clear cyclic or periodic trend (like a weekly or monthly spike pattern), which implies that orders are fairly uniformly distributed over time without strong seasonal impact—at least visually from this index-based view.

Random Variability:

The spikes are irregular and spread out, showing some unpredictability in large-quantity orders.
